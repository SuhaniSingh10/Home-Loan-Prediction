### Relation between "Loan_Status" and "Gender"

```python
# Creating a crosstab for Gender and Loan_Status
Gender = pd.crosstab(train["Gender"], train["Loan_Status"])

# Normalizing by the total count per Gender (row-wise)
Gender_percentage = Gender.div(Gender.sum(1).astype(float), axis=0)

# Plotting the stacked bar chart
ax = Gender_percentage.plot(kind="bar", stacked=True, figsize=(6, 4), color=["orange", "blue"])

# Adding labels and title
plt.xlabel("Gender")
plt.ylabel("Percentage")
plt.title("Loan Status by Gender")

# Moving the legend outside the plot
plt.legend(title='Loan Status', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjusting layout to make room for the legend
plt.tight_layout()

# Display the plot
plt.show()
```
